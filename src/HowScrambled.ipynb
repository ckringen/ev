## Complexity of Scrambling
Here we look at various ways to measure how different the scrambled sentence is from the target sentence.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("../data/data_cleaned.tsv", sep='\t')
print(data[:2])
print(list(data.columns.values))
```

```python
# get 50 responses for a single sentence (restrict to  given subset)
# sentence 58 for each list(1:4)
s58_1 = data.ix[(data['Sentence Number']==58) & (data['List Number']==1)]
s58_2 = data.ix[(data['Sentence Number']==58) & (data['List Number']==2)]
s58_3 = data.ix[(data['Sentence Number']==58) & (data['List Number']==3)]
s58_4 = data.ix[(data['Sentence Number']==58) & (data['List Number']==4)]

print(s58_1.iloc[0]['Target Sentence'], ":", s58_1.iloc[0]['Scrambled Sentence'])
print(s58_2.iloc[0]['Target Sentence'],":", s58_1.iloc[0]['Scrambled Sentence'])
print(s58_3.iloc[0]['Target Sentence'],":", s58_1.iloc[0]['Scrambled Sentence'])
print(s58_4.iloc[0]['Target Sentence'],":", s58_1.iloc[0]['Scrambled Sentence'])
```

```python
# get 50 responses for a single sentence (restrict to  given subset)
# sentence 58 for each list(1:4)
s58_1 = data.ix[(data['Sentence Number']==13) & (data['List Number']==1)]
s58_2 = data.ix[(data['Sentence Number']==13) & (data['List Number']==2)]
s58_3 = data.ix[(data['Sentence Number']==13) & (data['List Number']==3)]
s58_4 = data.ix[(data['Sentence Number']==13) & (data['List Number']==4)]

print(s58_1.iloc[0]['Target Sentence'], ":", s58_1.iloc[0]['Scrambled Sentence'])
print(s58_2.iloc[0]['Target Sentence'],":", s58_1.iloc[0]['Scrambled Sentence'])
print(s58_3.iloc[0]['Target Sentence'],":", s58_1.iloc[0]['Scrambled Sentence'])
print(s58_4.iloc[0]['Target Sentence'],":", s58_1.iloc[0]['Scrambled Sentence'])

fig = plt.figure(figsize=(18,12))
plt.subplot(311)
x = plt.hist(np.array(s58_1['editDist_scr_tgt']))
y = plt.hist(np.array(s58_2['editDist_scr_tgt']))
z = plt.hist(np.array(s58_3['editDist_scr_tgt']))
w = plt.hist(np.array(s58_4['editDist_scr_tgt']))
plt.legend(('List_1','List_2','List_3','List_4'))
plt.title("Average Edit Distance")

plt.subplot(312)
x = plt.hist(np.array(s58_1['avgBigramDist']))
y = plt.hist(np.array(s58_2['avgBigramDist']))
z = plt.hist(np.array(s58_3['avgBigramDist']))
w = plt.hist(np.array(s58_4['avgBigramDist']))
plt.legend(('List_1','List_2','List_3','List_4'))
plt.title("Average Ratio of Words splitting up Bigrams (by number of bigrams)")

plt.subplot(313)
x = plt.hist(np.array(s58_1['propMatchBigrams']))
y = plt.hist(np.array(s58_2['propMatchBigrams']))
z = plt.hist(np.array(s58_3['propMatchBigrams']))
w = plt.hist(np.array(s58_4['propMatchBigrams']))
plt.legend(('List_1','List_2','List_3','List_4'))
plt.title("Proportion of Matching Bigrams")

```
